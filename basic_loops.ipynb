# Basic Python Loops

```python
x = [1,2, 3, 4, 5]
for item in x:
    print(item)
```

```python
for i, item in enumerate(x):
    print(f"Index: {i}, Item: {item}")
```

```python
for i in range(len(x)):
    print(x[i])
```

## Control Structures

```python
x = [545, 2, 215, 3,48, 0.5]
for item in x:
    if item == 2:
        print(item)
```

```python
x = [545, 2, 215, 3,48, 0.5]
for item in x:
    if item == 2:
        continue
    else:
        print(item)
```

```python
x = [545, 2, 215, 3,48, 0.5]
for item in x:
    if item == 2:
        break
    else:
        print(item)
```
